# Crime in Each Neighborhood

Crime, as in safety, is often a large factor affecting many other things in the ecosystem of a neighborhood. Crime rates give lots of insight into the overall quality of life in a neighborhood. To determine the *best* neighborhood in Pittsburgh, you must also determine the safest. Here, we have a data set detailing the incidents reported by the City of Pittsburgh Police during the past seven years. 

### Let's breakdown this data!

Each entry in this data set is an incident. We have lots of information on each incident, such as the time, exact location, neighborhood, hierarchy of offense, etc. For the purpose of finding the *best*, safest neighborhood, we will only need to look at each incident and the neighborhood it occured in. Let's get rid of all of the other, unnecessary information.

```python
import pandas as pd

raw_data = pd.read_csv("crime.csv",  index_col="_id")

data = raw_data.drop(columns=['X', 'Y', 'INCIDENTTIME', 'PUBLIC_WORKS_DIVISION', 'HIERARCHY', 'COUNCIL_DISTRICT', 'INCIDENTTRACT', 'OFFENSES', 'INCIDENTZONE', 'INCIDENTLOCATION', 'CLEAREDFLAG', 'INCIDENTHIERARCHYDESC', 'CCR', 'PK'])
data.head(10)
```

<br> Now, we've condensed this data to allow us to focus on the neighborhood that each incident occured in. Next, let's see how many incidents have occured in each neighborhood and sort that list from least to greatest. 

```python
incident_count = data['INCIDENTNEIGHBORHOOD'].value_counts().sort_values(ascending=True)
incident_count
```

<br> These counts are telling us the total amount of recorded incidents in that neighborhood over the **past seven years**. This is a pretty incomparable statistic, so let's look at the average in each neighborhood **per year**. 

```python
per_year = incident_count/7
per_year
```

<br> Now that we have easily comparable data, we can base our metric off of it. First, let's look at it in a more digestible way: a graph! The bar graph below shows each neighborhood and the amount of incidents reported. 

```python
incidents_graph = per_year.head(98).plot.bar(x='INCIDENTNEIGHBORHOOD', y=per_year, figsize=(15,10))
```

From this bar graph we can see the true distribution of incidents per year visually, ranging from 14 to almost 2752 crimes.

<br> Although we already have this data sorted in ascending order, showing us the neighborhoods with the least crime, let's find a constant statistic to compare each value with to further give us an idea of how crime-free each neighborhood really is. 

Pennsylvania has 2,559 towns/neighborhoods and on average 247,190 crimes per year in total, meaning that the average number of crimes per neighborhood would be:

```python
247190/2559 
```

The average number of crimes per neighborhood is about 97 crimes per neighborhood in Pittsburgh. 

<br> Now that we've figured out a good comparison value, let's look back at our data for incidents per neighborhood per year. To figure out the *best* neighborhood in Pittsburgh, we want the neighborhoods with the least crime. Therefore, let's grab the first 20 neighborhoods and their incidents per year value from the list. 

```python
per_year.head(20)
```

<br> Since we sorted this list in ascending order, its first 20 values will be the neighborhoods with the smallest amount of incidents per year. Most of these values are less than the state average of 97 crimes per year. Only Esplen, Glen Hazel, and Central Northside have incident per year values greater than the state average. However, compared to the other Pittsburgh neigborhoods, these three still have very low incident per year values. 

### In Conclusion...

Looking back at the data of incidents per neighborhood per year above, we can determine that the Top 5 "safest" neighborhoods are Mt. Oliver Boro, Mt. Oliver Neighborhood, Outside County, Troy Hill-Herrs Island, and Ridgemont. Their incidents per year are the lowest out of all of the Pittsburgh neighborhoods in this data set, and are also the lowest compared to the Pennsylvania average crimes per year.

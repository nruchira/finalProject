# Education By Neighborhood
---

Often times a neighborhood or area can be judged based upon the enrollemnt and preformance of the students attending the local schools. The amount of individualized Learning Plans, or IEPs, often help show the accomodation level of a school - which often times acts as a reflection of the quality neighborhood where the students live. The more accomadating a school is, the better the neighborhood is. 

We will be using the IEP by Neighborhood 2020-21 school year reported by Pittsburgh Public Schools

```python
import pandas as pd

original_data = pd.read_csv("neighborhood_iep.csv",index_col="neighborhoods")
original_data.head(10)
```

Above is shows the first 10 schools listed in our original dataset, displayed in alphabetical order. Now that we have seen that, it is time to sort our data.

---

First, we will look at the top ten school districts/neighborhoods with the highest number of enrollments.

```python
total_enrolment = original_data.drop(columns=['gifted','other_iep', 'not_iep_or_gifted'])
total_enrolment.head(10)
```

Above is the chart that removes the extranious data for our current venture, unsorted. 

Bellow is this chart sorted from highest to lowest (including all the data, not just the first 10 samples). 

```python
sorted_enrollment = total_enrolment['total_enrollment_k_12'].sort_values(ascending=False)
print("All Neighborhoods sorted: \n")

sorted_enrollment
```

This is a bit clunky, so for the sake of visibility lets reduce this down to just the top 10

```python
print("Top Ten Neighborhoods:\n")
sorted_enrollment.head(10)
```

This view is a lot nicer, but there is more that we can do! 

#### To view this data better, bellow I have created a bar chart/ bar graph showing the spread.

```python
graph_of_enrollment = sorted_enrollment.plot.bar(x='neighborhoods', y='total_enrollment_k_12', figsize=(15, 10))
```

Because we've used a bar chart (that has been sorted, versus a ) not a lot can be said of the shape of the data, but just by glancing at this graph we can clearly see that the top 3 schools in regards to enrollment have enrollment numbers that are drastically larger than the other neighborhoods. 

For a better look at our top ten schools in regards to enrollment, see the bar chart bellow.

```python
graph_of_enrollment = sorted_enrollment.head(10).plot.bar(x='neighborhoods', y='total_enrollment_k_12', figsize=(15, 10))
```

#### Looking at all of this data, we can't actually come to a solid conclusion as to which school overall has the best education system in regards to IEPs in place. Another issue is that simply looking at the values based upon numerical value does not regard the size of each school district. Instead, we should look the the percentage of IEPs, both gifted and individualized, in regards to school population. The districts with the largest percentage is the school with the best education system.

---

To do what is mentioned above, what we are going to do is first create a dataframe containing the total number of IEPs in each district

```python
total_ieps = ((original_data['gifted'] + original_data['other_iep']))
total_ieps
```

Next, we will create another data frame that contains the percantage of the iep enrollments out of the total student enrolments in each neighborhood 

```python
percentage_df = total_ieps/total_enrolment['total_enrollment_k_12'] *100
percentage_df
```

Now, we will sort this data to find the schools top ten schools with the highest percentages of both Gifted and Individualized IEPs.

```python
sorted_percentages = percentage_df.sort_values(ascending=False)
sorted_percentages.head(10)
```

And once more, for ease of viewing, we will include a bar graph as well

```python
percentage = sorted_percentages.head(10).plot.bar(x='neighborhoods', y=sorted_percentages, figsize=(15, 10))
```
